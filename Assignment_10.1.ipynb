```python
import pandas as pd
import numpy as  np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
df.head()
```

## 1. Delete unnamed columns

```python
## drop unnamed column
df.drop(columns= 'Unnamed: 0', inplace=True)
df.head()
```

## 2. Show the distribution of male and female

```python
import matplotlib as plt
print(" Distribution across Male and Female")
print(df.Gender.value_counts(normalize=False))
df.Gender.value_counts(normalize=True).plot.pie()
```

## 3. Show the top 5 most preferred names

```python
#df.Name.value_counts(ascending=False)
df.groupby('Name').Count.sum().sort_values(ascending=False).head()
```

## 4. What is the median name occurence in the dataset

```python
df_Names = pd.DataFrame(df.groupby('Name').Count.sum().sort_values(ascending=False))
print("The median name occurence in the dataset is as follows:")
df_Names[df_Names.Count == df_Names.Count.median()]
```
